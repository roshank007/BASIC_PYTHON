```python
#IMPORTING LIBRARIES 
from gtts import gTTS
import os
```

```python
#DEFINE THE LANGUAGE AND ASSIGNING FILE INTO VARIABLE
audio = 'speech.mp3'
language = 'en'
```

```python
#CALLING gTTS, here can replace text with your text.
sp = gTTS(text = 'Welcome to python',
         lang = language,
         slow  = False)
sp.save(audio)
```
